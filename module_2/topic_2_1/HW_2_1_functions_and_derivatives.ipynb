### Домашні завдання на відпрацювання поняття функції та її похідної

Необхідні імпорти:

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
x = sp.symbols('x')
f1 = 10 * x - 5
df1_s = sp.diff(f1, x)
print(f"Отримано: {df1_s} похідна функції {f1}")
```

```python
x = np.linspace(0, 10, 100)
f1 = 10 * x - 5
df = np.full_like(x, 10)

plt.figure(figsize=(10, 5))
plt.plot(x, f1, label='f(x) = 10x - 5', color='blue')
plt.plot(x, df, label="f'(x) = 10", color='green',)


plt.title('Графіки функції та її похідної')
plt.xlabel('x - Кількість годин навчання')
plt.ylabel('y - Оцінка')

plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.legend()
plt.grid()

plt.show()
```

**Висновок**:

Функція $ f(x) = 10x - 5 $ є лінійною функцією однієї змінної виду $ f(x) = kx + b $.
Тобто значення оцінки зростає пропорційно до збільшення значення часу навчання.
 Значення $ k $ - кутового коефіцієнту дорівнює 10, що відповідно
 показує зміну відносно $ y $ по відношенню до зміни $ x $. Тобто що кожна година навчання нам збільшує оцінку на 10.

Маємо похідну такої функції
 $ f(x) = kx + b $; $ f'(x) = k$.

 Для нашого випадку
$ f(x) = 10x - 5 $; $ f'(x) = 10 $

 $ f'(x) = 10 $ - стале значення, вказує на те, що швидкість зміни функції залишається сталою.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')
f2 = sp.sin(x)
df2_s = sp.diff(f2, x)
print(f"Отримано: {df2_s} похідна функції {f2}")
```

```python
x = np.linspace(0, 20, 100)
f1 = np.sin(x)
df = np.cos(x)

plt.figure(figsize=(10, 5))
plt.plot(x, f1, label='f(x) = sin(x)', color='blue')
plt.plot(x, df, label="f'(x) = cos(x)", color='green',)


plt.title('Графіки функції та її похідної')
plt.xlabel('x')
plt.ylabel('y')

plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.legend()
plt.grid()

plt.show()
```

```python
x = sp.symbols('x')
f3 = sp.ln(x)
df3_s= sp.diff(f3, x)
print(f"Отримано: {df3_s} похідна функції {f3}")
```

```python
x = np.linspace(0.1, 10, 100)
f1 = np.log(x)
df = 1 / x

plt.figure(figsize=(10, 5))
plt.plot(x, f1, label='f(x) = log(x)', color='blue')
plt.plot(x, df, label="f'(x) = 1/x", color='green',)


plt.title('Графіки функції та її похідної')
plt.xlabel('x')
plt.ylabel('y')

plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.legend()
plt.grid()

plt.show()
```

**Висновок**:



*   $ f(x) = \sin(x) $:

Похідна $ f'(x) = \cos(x) $.

Функція $ f(x) $ визначена на всій числовій прямій та має множину значень $[-1; 1]$ та період $2 π$.
Бачимо, що в точках екстремумів маємо значення $ f'(x) = 0 $, це є точками де функція $ f(x) $ має дотичну, що буде розташована горизонатально
Бачимо, що в точках де $ f(x) $ збільшується, значення $ f'(x) > 0 $ а де зменшується  $ f'(x) < 0 $

*   $ f(x) = \ln(x) $:  

Похідна $ f'(x) = 1/x $.

Функція $ f(x) $ визначена на $ (0; -∞) $ та має множину значень $ (-∞; ∞) $.
Бачимо, що функція зростає, але швидкість росту зменшується зі збільшенням значень $ x $.







3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')
f4 = x ** 2 - 5 * x + 1
df4_s = sp.diff(f4, x)
print(f"Отримано: {df4_s} похідна функції {f4}")
```

```python
x = np.linspace(0, 100, 100)
f1 = x ** 2 - 5 * x + 1
df = 2*x - 5

plt.figure(figsize=(20, 5))
plt.plot(x, f1, label='f(x) = x^2 - 5x + 1', color='blue')
plt.plot(x, df, label="f'(x) = 2x - 5", color='green',)


plt.title('Графіки функції та її похідної')
plt.xlabel('x')
plt.ylabel('y')

plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.legend()
plt.grid()

plt.show()
```

```python
x = np.linspace(0, 5, 15)
f1 = x ** 2 - 5 * x + 1
df = 2*x - 5

x_min = 2.5
f_min = x_min**2 - 5*x_min + 1

plt.figure(figsize=(20, 5))
plt.plot(x, f1, label='f(x) = x^2 - 5x + 1', color='blue')
plt.plot(x, df, label="f'(x) = 2x - 5", color='green',)

plt.scatter(x_min, f_min, color='red', label='Мінімум функції', zorder=5)

plt.title('Графіки функції та її похідної')
plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.grid()

plt.show()
```

**Висновок**:

Функція $ f(x) = x^2 - 5x + 1 $ є квадратичною, має похідну $ f'(x) = 2x + 5$, що показує швидкість втрат залежно від обсягу виробництва, з мінімумом у точці $ x = 2.5 $.

Відповідно до графіка бачимо зменшення значень $ y $ для $ x < 2.5 $, та збільшуватися для $ x > 2.5 $.
